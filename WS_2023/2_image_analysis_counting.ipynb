## Setup

```python
import skimage as ski
import matplotlib.pyplot as plt
```

## Read Image

```python
img = ski.io.imread("https://figshare.com/ndownloader/files/34498448")
```

```python
plt.imshow(img)
```

```python
img.shape
```

## Convert to gray

```python
gray_img = ski.color.rgb2gray(img)
```

```python
gray_img.shape
```

```python
gray_img.min(), gray_img.max()
```

```python
plt.imshow(gray_img, cmap="gray")
```

## Draw histogram

```python
gray_img.shape
```

```python
gray_img.flatten().shape
```

```python
plt.hist(gray_img.flatten(), bins=256, range=(0,1));
```

## Add blurring

```python
blurred_img = ski.filters.gaussian(gray_img, sigma=1)
```

```python
plt.imshow(blurred_img, cmap="gray")
```

```python
plt.hist(blurred_img.flatten(), bins=256, range=(0,1));
```

## Create mask

```python
mask = blurred_img < 0.2
```

```python
mask.shape
```

```python
plt.imshow(mask, cmap="gray")
```

## Count collonies

```python
labeled_img, count = ski.measure.label(mask, return_num=True)
```

```python
count
```

## Task

Create a function `analyse_image`, that takes a filename as an argument and performs the analysis steps as above. You don't have to `imshow` all the images, just `imshow` the original image. Don't plot the histograms, take `0.2` as a fix threshold. The function should return the number of colonies.

```python
def analyse_image(filename):
  _, axs = plt.subplots(1,2)
  img = ski.io.imread(filename)
  axs[0].imshow(img)
  gray_img = ski.color.rgb2gray(img)
  blurred_img = ski.filters.gaussian(gray_img, sigma=1)
  mask = blurred_img < 0.2
  axs[1].imshow(mask, cmap="gray")
  _, count = ski.measure.label(mask, return_num=True)
  return count
```

```python
analyse_image("https://figshare.com/ndownloader/files/34498448")
```

```python
analyse_image("https://figshare.com/ndownloader/files/34498442")
```

```python
analyse_image("https://figshare.com/ndownloader/files/34498445")
```
